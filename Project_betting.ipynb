```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

path = "open-data-donnees-marche-des-jah-en-ligne-2023.csv"
raw = pd.read_csv(path, encoding="ISO-8859-1", sep=";")

raw = raw.drop(columns=['Unnamed: 15'])
raw = raw.drop(raw.index[26:43])

raw.head(100)
```

1. Clean years info
2. Transpose the dataset
3. Define types per column

```python
columns = raw.columns
renames = dict()

for column in columns:
    renames[column] = column.replace("Au ", "").replace("31/12/", "")

raw_renamed = raw.rename(columns=renames)

raw_transposed = raw_renamed.transpose()

df = raw_transposed.drop(raw_transposed.index[0])

df.columns = [
    #"Annee",
    "Nombre d'operateurs",
    "Nombre d'agrements",
    "Nombre agrements paris sportifs",
    "Nombre agrements paris hippiques",
    "Nombre agrements poker",
    "Nombre de comptes joueurs actifs total (en milliers)",
    "Nombre CJA Paris sportifs (en milliers)",
    "Nombre CJA Paris hippiques (en milliers)",
    "Nombre CJA Poker (en milliers)",
    "Nombre de joueurs uniques",
    "Mises paris sportifs (en M€)",
    "PBJ paris sportifs (en M€)",
    "Mises paris hippiques (en M€)",
    "PBJ paris hippiques (en M€)",
    "PBJ poker (en M€)",
    "Budget marketing médias (en M€)",
    "Part hommes PS",
    "Part femmes PS",
    "Part hommes PH",
    "Part femmes PH",
    "Part hommes PO",
    "Part femmes PO",
    "Part mises football",
    "Part mises tennis",
    "Part mises basketball",
    "Part mises rugby"
]

df.head(100)
```

```python
def french_to_english_decimal(number):
    if isinstance(number, str):
        return number.replace(',', '.')
    else:
        return number
df = df.map(french_to_english_decimal)
display(df)
```

```python

def is_percentage(value):
    return isinstance(value, str) and value.endswith('%')
df_is_percentage = df.map(is_percentage)


percentage_cols = df_is_percentage.any()
percentage_cols = percentage_cols[percentage_cols].index.tolist()

for col in percentage_cols:
    df[col] = df[col].str.rstrip('%').astype(float) / 100
    
df = df.astype(float)

print(df)
```

```python
df = df.apply(lambda col: col.fillna(col.min()))
```
